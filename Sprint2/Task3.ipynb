### Вариант задания №1


<h2 style="color:DodgerBlue">Описание проекта:</h2>

----

Создать базовый класс Person в C#, который будет представлять информацию о
человеке. На основе этого класса разработать 2-3 производных класса,
демонстрирующих принципы наследования и полиморфизма. В каждом из классов
должны быть реализованы новые атрибуты и методы, а также переопределены
некоторые методы базового класса для демонстрации полиморфизма.

Требования к базовому классу Person:
• Атрибуты: Имя (Name), Возраст (Age), Пол (Gender).
• Методы:
o GetInfo(): метод для получения информации о человеке в виде строки.
o SayHello(): метод для вывода приветствия от человека.

Требования к производным классам:
1. Студент (Student): Должен содержать дополнительные атрибуты, такие как
Университет (University) и Курс (Course). Метод SayHello() должен быть
переопределен для добавления информации о курсе при приветствии.
2. Работник (Employee): Должен содержать дополнительные атрибуты, такие
как Компания (Company) и Зарплата (Salary). Метод GetInfo() должен быть
переопределен для включения информации о компании и зарплате.
3. Преподаватель (Teacher) (если требуется третий класс): Должен содержать
дополнительные атрибуты, такие как Предмет (Subject) и Стаж работы
(Experience). Метод SayHello() должен быть переопределен для добавления
информации о преподаваемом предмете при приветствии.


#### Дополнительное задание
Добавьте к сущестующим классам (базовыму и производным 3-4 атрибута и метода) и реализуйте простое, сложное и множественное наследование

```
public interface IPerson {
    string Name{get;set;}
    int Age{get;set;}
    string Gender{get;set;}
}

public class Person : IPerson
{
    public string Name { get; set; }
    public int Age { get; set; }
    private string _gender;
    private static readonly string[] Genders = {"Женский", "женский", "Мужской", "мужской","ж","м","Ж","М"};
    //выше константа для гендера
    public string Gender
    {   get { return _gender; }
        set { if (Genders.Contains(value))
            {_gender = value;}
            else{throw new ArgumentException("Недопустимое значение пола");}}}
    public Person(string name, int age, string gender)
    { Name = name; Age = age; Gender = gender; // используется паблик переменная, а не приват
    }
    public virtual string GetInfo()
    { return($"{Name},{Age},{Gender}");
    }
      public virtual void SayHello()
    {
        Console.WriteLine($"Привет! Меня зовут {Name}. Мой возраст {Age}. Мой пол {Gender}");
        }
    protected void CheckAge(int Age){
        if (Age>=0 && Age<=17) {
            Console.WriteLine($"Человек с именем {Name} несовершеннолетний(яя), так как ему(ей) {Age} лет(годов)");
        }
        else if (Age>=18 && Age<=44)
        {
             Console.WriteLine($"Человек с именем {Name} совершеннолетний(яя) и является молодым(ой), так как ему(ей) {Age} лет(годов)");
        }
        else if (Age>=44 && Age<=60)
        {
             Console.WriteLine($"Человек с именем {Name} среднего возраста, так как ему(ей) {Age} лет(годов)");
        }
        else if (Age>=60 && Age<=120)
        {
             Console.WriteLine($"{Name} является пожилым, так как ему(ей) {Age} лет(годов)");
        }
        else{
            throw new ArgumentException("Недопустимое значение возраста");
        }}}
class Student : Person
 { 
    public string University{get;set;}
    private int _course;
    public int Course {
            get { return _course; }
            set{
                if ((value>0)&&(value<8)){
                    _course = value;
                }
                else{throw new ArgumentException("Ошибка. Проверьте цифру курса");}
            }}
    
    public static int countStudent{get; private set;}=0; 

    public Student(string name, int age, string gender, string university, int course): base(name, age, gender)
    {University=university;
        Course=course;
        countStudent++;
    }

     public override string GetInfo()
    {
        return (base.GetInfo()+$"{University},{Course}");//- вызов базового метода
        //return($"{University}, {Course}");
    }
    
     public new void SayHello()
    {
        Console.WriteLine($"Привет! Меня зовут {Name}. Мой возраст {Age}. Мой пол {Gender}. Я учусь в \"{University}\" на {Course} курсе.");
         Console.WriteLine($"По номеру я: {countStudent}-ый(ая) студент(ка)");
         
    }   
}
class Employee: Person{
 public string Company{get;set;}
    public double Salary {get;set;}
    public static int countEmployee{get; private set;}=0; 
    public Employee(string name, int age, string gender, string company, double salary): base(name, age, gender)
    {
        Company=company;
        Salary=salary;
        countEmployee++;
    }

     public override string GetInfo()
    {  return (base.GetInfo()+$"{Company},{Salary}");
    }
    public void CheckAgeEmployee (){
        CheckAge(Age);}    
    public new void SayHello()
    {
        Console.WriteLine($"Привет! Меня зовут {Name}. Мой возраст {Age}. Я работаю в компании \"{Company}\", моя зарплата составляет {Salary} рубля(ей).");
        CheckAgeEmployee();
    }

    
}
class Teacher: Employee{
 public string Subject{get;set;}
    public double Experience {get;set;}
    public static int countTeacher{get; private set;}=0; 
    public Teacher(string name, int age, string gender, string company, double salary,string subject,double experience): base(name, age, gender,company,salary)
    {
        Subject=subject;
        Experience=experience;
        countTeacher++;
    }
     public override string GetInfo(){
          return (base.GetInfo()+$"{Subject},{Experience}");
    }
    public void CheckAgeTeacher (){
        CheckAge(Age);
    }    
    public new void SayHello(){
        Console.WriteLine($"Привет! Меня зовут {Name}. Мой возраст {Age}. Я работаю учителем в \"{Company}\", моя зарплата составляет {Salary} рубля(ей).");
         Console.WriteLine($"Я преподаю учебный предмет \"{Subject}\" уже в течение {Experience} лет(года)");
        CheckAgeTeacher();
    }}

Person fly = new Person ("Григорий", 44, "мужской"); Console.WriteLine("Введенные даные: "+fly.GetInfo());
fly.SayHello();  Console.WriteLine();
Student cat = new Student ("Евгения", 23, "Женский", "ТИУ", 2); Console.WriteLine("Введенные даные: "+cat.GetInfo());
cat.SayHello(); Console.WriteLine();
Employee dog = new Employee ("Денис", 43, "Мужской", "Строительный двор", 23050.48);
dog.SayHello();
Console.WriteLine();
Teacher teacher = new Teacher("Ольга",75,"женский","Школа №44",46660,"История", 6);
teacher.SayHello(); teacher.GetInfo();
```
